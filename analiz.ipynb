```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
```

### Анализ 1: «Статистика количества персонажей по цвету их волос»

**Что показано:**
Распределение женских аниме-персонажей по цвету волос.

**Ключевые данные:**
Самые распространённые цвета: чёрные, фиолетовые, каштановые, розовые.

Разнообразие внешности - система предлагает широкий выбор, удовлетворяющий разные эстетические предпочтения пользователей (от классических до экстравагантных образов).

Акцент на фэнтези-эстетике - преобладание неестественных цветов подчёркивает «анимешную» стилистику, что соответствует ожиданиям целевой аудитории.

Персонализация рекомендаций - можно внедрить фильтр или алгоритм подбора персонажа по любимому цвету волос пользователя.

```python
np.random.seed(10)

with open("characters.json", "r", encoding="utf-8") as f:
    json_data = json.load(f) 

df = pd.DataFrame(json_data["female_anime_characters"])

print("Статиcтика количества персонажей по цвету волос:")

hair_color_count = df["hair_color"].value_counts()

i = 1
for color, count in hair_color_count.items():
    percent = (count / len(df)) * 100
    print(f"{i}) {color} волосы - {count} персонажей ({int(percent)}%)")
    i += 1
    
plt.figure(figsize=(16, 8))
hair_colors = ["Чёрные",
    "Фиолетовые",
    "Каштановые",
    "Розовые",
    "Голубые",
    "Серебряные",
    "Рыжие",
    "Зелёные",
    "Блонд",
    "Разноцветные",
    "Бирюзовые",
    "Синие",
    ]

diagram2 = plt.bar(hair_colors, hair_color_count, color="pink", alpha= 0.8)
plt.xlabel("Цвет волос персонажа")
plt.ylabel("Количество персонажей")
plt.title("Статистика количества персонажей по цвету их волос", fontweight = 14)

plt.tight_layout()
plt.show()
```

### Анализ 2: «Статистика количества персонажей по их возрасту»

**Что показано:**
Распределение персонажей по возрастным группам.

**Ключевые данные:**

Абсолютное большинство персонажей относятся к группам 16-18, 18-20, 20-25 лет.

Минимум персонажей в категориях 40-60 и 100+ (последние, скорее, фэнтезийные долгожители).

**Целевая аудитория** - молодые персонажи соответствуют ожиданиям основной аудитории таких проектов (подростки и молодые взрослые).

Эмоциональный резонанс - возраст 16-25 лет ассоциируется с юностью, романтикой, становлением личности, что идеально для сюжетов свиданий.

```python
print(" "  * 20)
print("Статистика персонажей по возрасту:")

age_group_count = df["age_group"].value_counts()

i = 1
for i,  (age_group, count) in enumerate(age_group_count.items(), 1):
    print(f"{i}) {age_group} - {count} персонажей")
    
plt.figure(figsize=(10, 8))
age_groups = ["16-18",
    "18-20",
    "20-25",
    "25-35",
    "40-60",
    "100+",
    ]

diagram = plt.bar(age_groups, age_group_count, color="black", alpha= 0.8)
plt.xlabel("Возраст персонажа")
plt.ylabel("Количество персонажей")
plt.title("Статистика количества персонажей по их возрасту", fontweight = 14)

plt.tight_layout()
plt.show()
```

### Анализ 3: «Статистика популярности жанров аниме-персонажей»

**Что показано:**
Рейтинг популярности жанров, к которым относятся персонажи (данные сгенерированы случайно, но отражают типичную ситуацию).

**Ключевые данные:**
Высокая популярность у жанров: романтика, комедия, драма, повседневность, фэнтези.

Средние значения: сёнэн, сёдзё, мистика.

Ниже популярность у меха, спорт, самураи.

Ядро жанров - романтика, комедия и драма идеально подходят для сценариев свиданий, диалогов, развития отношений.

жанр определяет антураж, сюжетные линии и характер персонажа

**Сегментация пользователей:** можно рекомендовать персонажей по любимым жанрам пользователя.

```python
print(" " * 20)

print("Рейтинг популярности жанров(своя рандомная генерация данных):")
list_genres = []
for genres in df["genres"]:
    for genre in genres:
        list_genres.append(genre)
different_genres = list(set(list_genres))

genre_popularity = {}
for genre in different_genres:
    popularity = np.random.uniform(30, 95)
    genre_popularity[genre] = popularity
    
sorted_popularity_genres = sorted(genre_popularity.items(), key=lambda x: x[1], reverse=True)

for i,  (genre, popularity) in enumerate(sorted_popularity_genres, 1):
    print(f"{i}) {genre} - {popularity:.1f}%")
    
plt.figure(figsize=(10, 8))
genres_list = [item[0] for item in sorted_popularity_genres]
popularity_list = [item[1] for item in sorted_popularity_genres]

diagram3 = plt.bar(genres_list, popularity_list, color="blue", alpha= 0.8)
plt.xlabel("Жанры")
plt.ylabel("Популярность (%)")
plt.title("Статистика популярности жанров аниме-персонажей", fontweight = 14)
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
